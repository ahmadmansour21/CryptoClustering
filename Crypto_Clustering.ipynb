```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
numerical_columns = df_market_data.select_dtypes(include=['float64', 'int64']).columns.tolist()

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df_market_data[numerical_columns])

scaled_data
```

```python
df_market_data.columns
```

```python
# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=numerical_columns)
# Copy the crypto names from the original DataFrame
scaled_df['coin_id'] = df_market_data.index.values

# Set the coin_id column as index
scaled_df.set_index('coin_id', inplace=True)

# Display the scaled DataFrame
scaled_df.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for k in range(1, 12):
    k_model = KMeans(n_clusters=k, random_state=1) # 1. Create a KMeans model using the loop counter for the n_clusters
    k_model.fit(scaled_df) # 2. Fit the model to the data using `df_market_data_scaled`
    inertia.append(k_model.inertia_) # 3. Append the model.inertia_ to the inertia list

print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': list(range(1, 12)),  
    'inertia': inertia  
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    xticks=list(range(1, 12))  
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the graph and the inertia table, it appears that k = 3 is the best value. The drop from k=2 to k = 3 is not as significant as the drop from k = 3  to k = 4. Similarly, the change from k = 4 to k = 5 is much less significant than that of k = 2 to k = 3, rendering k = 3 the best option.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 3
kmeans_model = KMeans(n_clusters=best_k, random_state=1)

print(kmeans_model)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
cluster_labels = kmeans_model.predict(scaled_df)

# Print the resulting array of cluster values.
print("Cluster Labels:")
print(cluster_labels)
```

```python
# Create a copy of the scaled DataFrame
scaled_df_copy = scaled_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
scaled_df_copy['Predicted_Cluster'] = cluster_labels

# Display the copy of the scaled DataFrame
scaled_df_copy
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = scaled_df_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Predicted_Cluster",  # Color the points by the predicted clusters
    hover_cols=["crypto_name"],  # Add cryptocurrency name for identification
    title="Cryptocurrency Price Change Percentages",
    xlabel="24h Price Change Percentage",
    ylabel="7d Price Change Percentage"
)
scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(scaled_df)

pca_df = pd.DataFrame(pca_data, columns=["PCA1", "PCA2", "PCA3"])

# View the scaled PCA data
pca_df.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
explained_variance
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** In order to get the total variance of the three principal components, we would sum up their individual variances. So, 0.3719856 + 0.34700813 + 0.17603793 = roughly 0.895. This means that these components account for roughly 89.5% of variance.

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(pca_data, columns=["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original scaled DataFrame
pca_df['coin_id'] = scaled_df.index.values

# Set the coin_id column as index
pca_df.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
pca_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Create a KMeans model using the loop counter for n_clusters
    model = KMeans(n_clusters=k, random_state=1)
    
    # Fit the model to the data using `df_market_data_pca`
    model.fit(pca_df)
    
    # Append the model's inertia to the inertia list
    inertia.append(model.inertia_)

inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': list(range(1, 12)),  
    'inertia': inertia  
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data)

df_elbow_pca
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    xticks=k_values
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Based on both the graph and the table, it appears that the best value for k here is k = 4 for the same reasons as why it was k = 3 before. Determined using the elbow method.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Yes, the answer does differ as it was k = 3 previously but k = 4 now.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 4
kmeans_model = KMeans(n_clusters=best_k, random_state=1)

print(kmeans_model)
```

```python
# Fit the K-Means model using the PCA data
kmeans_model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
crypto_clusters = kmeans_model.predict(pca_df)

# Print the resulting array of cluster values
print(crypto_clusters)
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_predictions = pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_predictions['StockCluster'] = crypto_clusters  

# Display the copy of the scaled PCA DataFrame
df_pca_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
pca_scatter_plot = df_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="StockCluster",  
    hover_cols=["Ticker"],  
    title="Cryptocurrency Clusters Based on PCA",
    xlabel="Principal Component 1 (PC1)",
    ylabel="Principal Component 2 (PC2)"
)
pca_scatter_plot
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
composite_plot = df_elbow.hvplot.line(
    x='k',
    y='inertia',
    label='K-Means',
    color='blue'
) * df_elbow_pca.hvplot.line(
    x='k',
    y='inertia',
    label='Agglomerative Clustering',
    color='green'
)

# Add titles and labels
composite_plot.opts(
    title='Comparison of Elbow Curves',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    xticks=list(range(1, 12)),
    legend_position='top_left'
)

composite_plot
```
